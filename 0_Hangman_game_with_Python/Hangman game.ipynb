# Hangman game

Hangman is a guessing game where the player(s) try to guess the word by suggesting a letters.
Player is allowed a certain number of mistakes.
The game starts with a word where all letters are replaced with a `-` so the only information available to a player is a number of letters in a word.
**Start**: `------`
For each correctly guessed letter, a position of the letter is revealed (if there are multiple positions for a same letter, they are all revealed.
**Correct**: `-E--E-`
For each wrongly guessed letter, number of allowed mistakes decreases by 1.
**Wrong**: 'Number of allowed mistakes is now 5'

Name of the game comes from a drawing of a hanged stick figure. For every mistake a new element of a hanged figure is added.
If a player guesses a word before the last element of a figure is added, a player wins.
If a player doesn't guess a word before the last element of a figure is added, player loses.

##### Solution

```python
word = 'SECRET'
hidden_word = '-'*len(word)
mistake_count = 6
print(f'Hidden word is {hidden_word}')
while mistake_count > 0 and hidden_word.count('-') != 0:
    letter_count = 0
    index_count = 0
    print('Pick a letter:')
    picked_letter = input().capitalize()
    for letter in word:
        if letter == picked_letter:
            hidden_word = hidden_word[:index_count] + picked_letter + hidden_word[index_count+1:]
            letter_count += 1
        index_count += 1
    if word.count(picked_letter) > 0:
        print(f'CORRECT! The word contains the {picked_letter}')
        print(hidden_word)
    else:
        mistake_count -= 1
        print(f'WRONG! Number of mistakes left: {mistake_count}')
if mistake_count == 0:
    print('HANGED!!!')
else:
    print('CONGRATULATIONS!!!')
```

##### Solution

```python
from random import randrange
list_of_words = ['LEARNING','PYTHON','CREATING','GAME']
word = list_of_words[randrange(len(list_of_words))]
hidden_word = '-'*len(word)
print(f'Hidden word is: {hidden_word}')
mistake_count = 6
print('==============================================================')
while mistake_count > 0 and hidden_word.count('-') != 0:
    index_count = 0
    letter_count = 0

    print('Pick a letter:')
    picked_letter = input().capitalize()
    for letter in word:

        if(picked_letter == letter):
            hidden_word = hidden_word[:index_count] + picked_letter + hidden_word[index_count+1:]
            letter_count += 1
        index_count += 1
    if(letter_count == 0):
        mistake_count -= 1
        print(f'WRONG! Number of mistakes left: {mistake_count}')
        print(hangman[6 - mistake_count])
    else:
        print(f'CORRECT! There is a letter {picked_letter} in a secret word!')
        print(hidden_word)
    print('==============================================================')
if(mistake_count == 0):
    print('HANGED!!!')
else:
    print('CONGRATULATIONS!!!')
```

## Advanced solution with new features

```python
# Import necessary libraries
from random import randrange

tableflip = {
    0: '''
         ( )
         \|/
          |
         / \\
    ''',
    1: '''
         (ಠ_ಠ)
         \|/
          |
         / \\
    ''',
    2: '''
        ┌(ಠ_ಠ)┘
         \|/
          |
         / \\
    ''',
    3: '''
        ┌(ಠ_ಠ)┘
         \|/
        /|
         / \\
    ''',
    4: '''
        ┌(ಠ_ಠ)┘
         \|/
        /|\\
         / \\
    ''',
    5: '''
        ┌(ಠ_ಠ)┘
        ┌|/
        /|\\
         / \\
        Table
    ''',
    6: '''
        (╯ಠ_ಠ)╯︵ ┻━┻
        GAME OVER
        Table flipped!
    '''
}

# Difficulty Levels
difficulty_levels = {
    'easy': 8,
    'medium': 6,
    'hard': 4
}

# Ask the player to choose a difficulty level
print("Choose a difficulty level (easy, medium, hard):")
difficulty = input().lower()
while difficulty not in difficulty_levels:
    print("Invalid difficulty. Please choose 'easy', 'medium', or 'hard':")
    difficulty = input().lower()

mistake_count = difficulty_levels[difficulty]  # Set the number of allowed mistakes based on difficulty

# Initialization of word list and selection of a random word
list_of_words = {
    'easy': ["CAT", "DOG", "FISH"],
    'medium': ["PYTHON", "PRIMER", "ARRAY"],
    'hard': ["BOOTCAMP", "HANGMAN", "FUNCTION"]
}

word = list_of_words[difficulty][randrange(len(list_of_words[difficulty]))]  # Selecting a word based on difficulty
hidden_word = len(word) * "-"  # Creating a hidden version of the word

# Display the hidden word to the player
print(f"Hidden word is {hidden_word}")

# Separator for game's UI clarity
print('==============================================================')

# Game loop starts here
# Continue the loop as long as there are mistakes remaining and hidden letters to be guessed
while mistake_count > 0 and hidden_word.count("-") != 0:
    print("Pick a letter and press ENTER! (Type 'ESC' to quit)")
    picked_letter = input().capitalize()  # Capitalize the input for consistency

    # Check for quit command
    if picked_letter == "Esc": # As we are capitalizing the picked_letter, we have to check for "Esc" instead of "ESC"
        print("Game over. You've chosen to exit the game.")
        break  # Exit the game loop

    # Input validation: Check if the input is a single alphabetic character
    if not picked_letter.isalpha() or len(picked_letter) != 1:
        print("Invalid input. Please enter a single letter.")
        continue  # Skip the rest of the loop and prompt for input again

    index_count = 0  # Initialize index counter to track position in the word
    found_letter = False  # Flag to track if the picked letter is found in the word

    # Iterate through each letter in the word to check against the picked letter
    for letter in word:
        if letter == picked_letter:
            # If the picked letter matches the current letter, reveal it in the hidden_word
            hidden_word = hidden_word[:index_count] + picked_letter + hidden_word[index_count+1:]
            found_letter = True  # Set the flag to True since the letter was found
        index_count += 1  # Move to the next index position in the word

    # Check if the picked letter was found in the word
    if found_letter:
        print(f"CORRECT! The word indeed contains the letter '{picked_letter}'.")
        print(f"Hidden word is {hidden_word}")  # Show the updated hidden word
    else:
        # If the letter was not found, decrement the mistake count and inform the player
        mistake_count -= 1
        print(f"WRONG! Number of mistakes left: {mistake_count}")
        # Show a /tableflip figure based on mistakes
        print(tableflip[difficulty_levels[difficulty] - mistake_count])

    # Print a separator for better readability
    print('==============================================================')

# Game over conditions: Check if the game is won or lost
if mistake_count == 0:
    print("YOU'RE HANGED BUDDY!!")  # Player used up all mistakes without guessing the word
if picked_letter == "Esc":
    print("See you next time!") # Player chose to exit the game
else:
    print("CONGRATULATIONS!! You guessed the word!")  # Player successfully guessed the word


```
