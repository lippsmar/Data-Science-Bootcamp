## 1. Select all books from the table `titles` that contain the word "cooking" in their title

````python
titles.loc[titles['title'].str.contains('cooking', case = False)]
````

### Hint:

In SQL the syntax is:

```sql
SELECT title
FROM titles
WHERE title LIKE '%cooking%';
```

## 3. Select the full names (first and last name) of authors whose last name starts with "S"

````python
authors.loc[authors['au_lname'].str.startswith('S'), ['au_fname', 'au_lname']]
````

alternative solution

````python
# authors.loc[authors['au_lname'].str.contains('^S', regex=True), ['au_fname', 'au_lname']]
````


## 4. Select the name and address of all stores located in an Avenue (its address ends with "Ave.")

````python
stores.loc[stores['stor_address'].str.endswith('Ave.')]
````

alternative solution

````python
# stores.loc[stores['stor_address'].str.contains('Ave\.$', regex = True)]
````

### Hint:

In SQL the syntax is:

```sql
SELECT stor_name, stor_address
FROM stores
WHERE stor_address LIKE ("%Ave.");
```

## 5. Select the name and address of all stores located in an Avenue or in a Street (address ended in "St.")

````python
(
stores.loc[(stores['stor_address'].str.endswith('Ave.'))
          |(stores['stor_address'].str.endswith('St.')), ['stor_name', 'stor_address']]
)
````

### Hint:

In SQL the syntax is:

```sql
SELECT
	stor_name,
	stor_address
FROM
	stores
WHERE 
	stor_address LIKE '%St.' 
    OR stor_address LIKE '%Ave.';
```

## 6. Look at the "employee" table (select all columns to explore the raw data):
   Find a pattern that reveals whether an employee is Female or Male.
   Select all female employees.

````python
employee.loc[employee['emp_id'].str.endswith('F')]
````

alternative solution

````python
# employee.loc[employee['emp_id'].str.contains('F$', regex=True)]
````

### Hint:

In SQL the syntax is:

```sql
SELECT emp_id, fname,lname 
FROM employee 
WHERE emp_id LIKE ('%F');
```

## 7. Select the first and last names of all male employees whose name starts with "P".

````python
(
employee.loc[employee['emp_id'].str.endswith('M')
            & employee['fname'].str.startswith('P'), ['fname', 'lname']]
)
````

### Hint:

In SQL the syntax is:

```sql
SELECT
    emp_id,
    fname,
    lname
FROM
    employee
WHERE
    (fname LIKE 'P%' 
    OR lname LIKE 'P%') 
    AND emp_id LIKE '%M';
```
